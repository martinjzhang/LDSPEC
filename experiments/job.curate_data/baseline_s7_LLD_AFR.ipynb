```python
import pandas as pd
import numpy as np
import io
import gdreg

%load_ext autoreload
%autoreload 2
```

```python
PGEN_FILE = '/n/scratch3/users/j/jz286/imp_geno/ukb_imp_chr@_v3'
LD_FILE = '/n/groups/price/martin/data_GDREG/UKBimp_337K_MAF001/baseline_annot/LLD_AFR/1000G.AFR.QC.@.l2.ldscore.gz'
OUT_FILE = '/n/groups/price/martin/data_GDREG/UKBimp_337K_MAF001/baseline_annot/LLD_AFR/LLD_AFR.@.annot.gz'

for CHR in range(1, 23):
# for CHR in [22]:
    print('CHR%d' % CHR)
    df_snp_chr = gdreg.util.read_pgen(PGEN_FILE.replace('@', '%d' % CHR))['pvar']
    print('df_snp_chr', df_snp_chr.shape)

    df_ld = pd.read_csv(LD_FILE.replace("@", "%d"%CHR), sep='\t')
    temp_dic = {x:y for x,y in zip(df_ld['SNP'], df_ld['L2']) if y!=0}
    print('df_ld', df_ld.shape, 'MAF_Adj_LLD_AFR', len(temp_dic))
    
    df_annot = df_snp_chr[['CHR', 'SNP', 'BP']].copy()
    df_annot['AN:MAF_Adj_LLD_AFR'] = [temp_dic[x] if x in temp_dic else 1 for x in df_annot['SNP']]
    print('df_annot', df_annot.shape, 'Mapped MAF_Adj_LLD_AFR', (df_annot['AN:MAF_Adj_LLD_AFR']!=1).sum())
    gdreg.util.write_annot(df_annot, OUT_FILE.replace("@", "%d"%CHR))
```
