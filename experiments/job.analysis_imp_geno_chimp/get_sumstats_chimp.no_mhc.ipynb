```python
import pandas as pd
import numpy as np
import ldspec
import os
```

```python
PGEN_FILE = '/n/scratch3/users/j/jz286/imp_geno_chimp/ukb_imp_chr@_v3_chimp'
dic_data = {}
for CHR in range(1, 23):  # Check all 23 CHRs
    dic_data[CHR] = ldspec.util.read_pgen(PGEN_FILE.replace("@", "%s" % CHR))
    
dic_info = {}
for CHR in range(1, 23):
    dic_info.update({x:(y,z) for x,y,z in zip(
        dic_data[CHR]['pvar']['SNP'],
        dic_data[CHR]['pvar']['CHR'],
        dic_data[CHR]['pvar']['BP'],
    )})
```

```python
SUMSTATS_FILE = '/n/groups/price/martin/LDSPEC_data/UKBimp_337K_MAF001_chimp/sumstats/@.sumstats.gz'
SUMSTATS_FILE_OUT = '/n/groups/price/martin/LDSPEC_data/UKBimp_337K_MAF001_chimp/sumstats/@.nomhc.sumstats.gz'
TRAIT_LIST = pd.read_csv(
    '/n/groups/price/martin/LDSPEC_data/UKBB_trait/trait_list_all.txt', header=None
)[0]
print('TRAIT_LIST', len(TRAIT_LIST))
for i_trait,trait in enumerate(TRAIT_LIST):
    if os.path.exists(SUMSTATS_FILE_OUT.replace('@', trait)) is True:
        print('Finished', trait, i_trait)
        continue
    if os.path.exists(SUMSTATS_FILE.replace('@', trait)) is False:
        print('Missing', trait, i_trait)
        continue
    print('Computing', trait, i_trait)
    df_sumstats = pd.read_csv(SUMSTATS_FILE.replace('@', trait), sep="\t", index_col=None)
    ind_mhc = [
        (dic_info[x][0]==6) & (dic_info[x][1]>=25e6) & (dic_info[x][1]<=34e6) for x in df_sumstats['SNP']
    ]
    print('    df_sumstats', df_sumstats.shape)
    ind_mhc = np.array(ind_mhc, dtype=bool)
    df_sumstats = df_sumstats.loc[~ind_mhc]
    print('    df_sumstats', df_sumstats.shape)
    df_sumstats.to_csv(SUMSTATS_FILE_OUT.replace('@', trait), sep="\t", index=False)
```
